```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y,'ro--')# লাইন গ্রাফ আঁকে
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()         # গ্রাফ দেখায়
```

```python
# creating subplot

plt.subplot(2,2,1) # 2 row 2 column 1st plot
plt.plot(x,y,'ro--') # লাইন গ্রাফ আঁকে
plt.subplot(2,2,2) # 2 row 2 column 2nd plot
plt.plot(x,y,'bo--') # লাইন গ্রাফ আঁকে
plt.subplot(2,2,3) # 2 row 2 column 3rd plot
plt.plot(x,y,'go--') # লাইন গ্রাফ আঁকে
```

```python
import numpy as np
x=np.arange(0,10)
y=3*x+5
plt.title("matplotlib demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
# compute the x and y cocordinates for points on a sine curve 
x=np.arange(0, 3 * np.pi, 0.1)
y=np.sin(x)
plt.title("sine wave")

plt.plot(x,y)
plt.show()
```

```python
#subplot 
#compute the x and y coordinates for points on a sine curve
x=np.arange(0, 3 * np.pi, 0.1)
y_sine=np.sin(x)
y_cosine=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x, y_sine)
plt.title('Sine Wave')

plt.subplot(2,1,2)
plt.plot(x, y_cosine)
plt.title('Cosine Wave')
```

# Box plot using matplotlib

```python
data=np.random.normal(0, 1, 1000)  # Generate random data

#rectengular box plot

plt.boxplot(data, vert=True, patch_artist=True, notch=True)
plt.boxplot(data, vert=True, patch_artist=False, notch=True)
```

# piechart

```python
import matplotlib.pyplot as plt

# ডেটা
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]

# চার্ট আঁকা
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# শিরোনাম
plt.title("Fruit Distribution")

# সার্কেল নিশ্চিত করতে
plt.axis('equal')  # Pie যেন গোল দেখায়

# গ্রাফ দেখানো
plt.show()
```
