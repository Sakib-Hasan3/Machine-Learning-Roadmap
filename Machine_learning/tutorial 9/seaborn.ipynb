# Seaborn

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.load_dataset("tips")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
sns.displot(tips["total_bill"], kde=True);
plt.show()
```

```python
sns.displot(tips["total_bill"], kde=True,bins=30);
plt.show()
```

```python
sns.histplot(tips["total_bill"], kde=True, bins=30);
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde");
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='smoker')
plt.show()
```

```python
sns.rugplot(tips["total_bill"]);
plt.show()
```

```python
sns.displot(tips, x="total_bill", y="tip", hue="smoker", kind="kde")
plt.show()
```

```python
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, palette="Blues",estimator=sum)
plt.show()
```
