```python
import nltk
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
from nltk import word_tokenize, pos_tag

sentence = "The children are running fast in London."

tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

print("Tokens:", tokens)
print("\nPOS Tags:")
for word, tag in tags:
    print(word, "→", tag)
```
